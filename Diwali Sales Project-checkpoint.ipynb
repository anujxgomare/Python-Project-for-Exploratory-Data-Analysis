```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
    # to avoide encoding error, use 'unicode_escape'
```

```python
df.shape
```

```python
df.head(10)
```

```python
## Data Cleaning
```

```python
df.info()
```

```python
#drop unreleated/blank columns
df.drop(['Status' , 'unnamed1'], axis=1, inplace=True) #axis=1 means full row, inplace means to save the changes made
df.info()
```

```python
#check for NULL Values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
##Drop Null Values
df.dropna(inplace=True)
```

```python
df.shape  # after execution, 12 rows are deleted i.e null values deleted
df.info()
```

```python
# Change Data Type
df['Amount']=df['Amount'].astype('int')  #changed from float to int
df['Amount'].dtypes 
```

```python
## Rename column
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
#describe() methord returns description of data in the Dataframe(i.e. count,mean,std etc)
df.describe()
```

```python
#describe for specific column
df[['Age','Orders','Amount']].describe()
```

```python
##EXPLORATORY DATA ANALYSIS
```

```python
sns.countplot(x='Gender',data=df)
```

```python
# to get the count on bars
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:   
    ax.bar_label(bars)
```

```python
#sort the amount spend by m & w
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False )
```

```python
#purchasing is done by whom m or 5
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
## Freom above fgraph, most of the buyers are females and purchasing power of Female is more than men
```

```python
#AGE
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:   #this function is used to gve values on bar
    ax.bar_label(bars)
```

```python
## Total Amount vs Age Group
sales_age=df.groupby(['Age Group'], as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='rainbow')    #palette is used to change the color theme of bars
```

```python
##From above graphs most buyers are of Age Group 26-35
```

```python
#total no. of orders from top 10 states

sales_state=df.groupby(['State'], as_index=False).sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
## to diaply the axis data in visible way i.e avopiding the upper data clustering iun X asis
sales_state=df.groupby(['State'], as_index=False).sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})   
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# From the above graph, most of the orders are from Uttar Pradesh, Maharashtra & Karnataka
```

```python
## MARITAL STATUS
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# From above graph, purchasing power of married women is high
```

```python
#OCCUPATION

ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False).sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

```python
# From above graph,most of buyers are woking in IT, Aviation & HealthCare sector
```

```python
## Product Category
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False).sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

```python
# from above graph, most of products sold are from Food, Footware & Electronic catrgory
```

```python
#Conclusion :
#Married women age group 26-35 from Uttar Pradesh, Maharashtra, and Karnataka working in IT, Heathcare and Aviation are more likely to buy products from Food, Clothing & Electronic Category
```
